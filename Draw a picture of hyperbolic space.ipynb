```python
import numpy as np
import matplotlib.pyplot as plt

def plot_poincare_disk(num_geodesics=40):
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.axis('off')  

   
    boundary = plt.Circle((0, 0), 1, edgecolor='black', facecolor='none')
    ax.add_artist(boundary)

   
    def plot_geodesics(offset, n):
        for i in range(n):
            angle = 2 * np.pi * i / n
            center_x = offset * np.cos(angle)
            center_y = offset * np.sin(angle)
            radius = np.sqrt(center_x**2 + center_y**2 - 1)

            t = np.linspace(0, 2*np.pi, 400)
            x = center_x + radius * np.cos(t)
            y = center_y + radius * np.sin(t)

            mask = x**2 + y**2 <= 1
            plt.plot(x[mask], y[mask], 'k')  

    
    plot_geodesics(1.01, 40)

   
    plot_geodesics(1.06, 15)

   
    plot_geodesics(1.8, 5)


    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)
    plt.show()

plot_poincare_disk()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_elliptic_paraboloid():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

   
    x = np.linspace(-1, 1, 100)
    y = np.linspace(-1, 1, 100)
    x, y = np.meshgrid(x, y)

    
    # z = x**3 + y**2
    z = x**2 + y**2


    
    # ax.plot_surface(x, y, z, rstride=4, cstride=4, color='#84D44A', alpha=0.5)
    ax.plot_surface(x, y, z, rstride=4, cstride=4, cmap='viridis', alpha=0.8)

   
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    
    ax.set_axis_off()
    
    ax.view_init(elev=30, azim=45)
    
    plt.savefig('Lorenz.pdf', format='pdf')
    plt.show()

plot_elliptic_paraboloid()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_sphere():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = np.outer(np.cos(u), np.sin(v))
    y = np.outer(np.sin(u), np.sin(v))
    z = np.outer(np.ones(np.size(u)), np.cos(v))

    
    ax.plot_surface(x, y, z, rstride=4, cstride=4, cmap='viridis', alpha=0.8)

    
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    
    ax.set_axis_off()
    
    ax.view_init(elev=30, azim=45)

    plt.show()

plot_sphere()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to plot circles
def plot_circle(ax, center, radius):
    circle = plt.Circle(center, radius, edgecolor='white', facecolor='none', linewidth=0.5)
    ax.add_patch(circle)

# Recursive function to generate nested circles
def nested_circles(ax, center, radius, depth):
    if depth == 0:
        return
    plot_circle(ax, center, radius)
    new_radius = radius / 2
    angles = np.linspace(0, 2 * np.pi, 6, endpoint=False)
    for angle in angles:
        new_center = (center[0] + new_radius * np.cos(angle), center[1] + new_radius * np.sin(angle))
        nested_circles(ax, new_center, new_radius, depth - 1)

# Setting up the figure and axis
fig, ax = plt.subplots()
ax.set_aspect('equal')

# Main circle representing the boundary of the Poincaré disk
boundary = plt.Circle((0, 0), 1, edgecolor='black', facecolor='none', linewidth=1)
ax.add_patch(boundary)

# Generating nested circles within the Poincaré disk
nested_circles(ax, (0, 0), 1, 5)  # Adjust the depth value for more levels of nesting

# Adjusting the plot limits and background
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.set_facecolor('black')

# Hiding the axes
ax.axis('off')

plt.show()
```
